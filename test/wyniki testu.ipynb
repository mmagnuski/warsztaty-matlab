```python
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
wyniki = pd.read_excel('wyniki.xlsx')
wyniki.head()
```

```python
g = sns.kdeplot(wyniki['perc'], bw=0.2, shade=True, clip=(0,1))
g.set_xlim((0.38,1.1))
```

```python
max_pnt = wyniki.apply(max, axis=0)
mean_pnt = wyniki.apply(np.mean)
diffic = max_pnt - mean_pnt
diffic = diffic.drop(['ID', 'sum', 'max', 'perc'])
diffic.plot(figsize=(15,4), title='trudnoÅ›Ä‡ pytaÅ„');
```

```python
all_points = wyniki.drop(['ID', 'sum', 'max', 'perc'], axis=1)
all_points.transpose().plot(figsize=(14,6), legend=False)
```
